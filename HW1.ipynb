```python
# Задание 1
p = '   _~_    '
t = '  (o o)   '
i = ' /  V  \  '
c = '/(  _  )\ '
a = '  ^^ ^^   '
n = int(input('Сколько пингвинов ты хочешь увидеть? Введи число от 1 до 9! '))
if n > 9:
    print("Много! Все не влезут")
else:
    if n < 1:
        print("Все пингвины разбежались!")
    else:
        print (p*n)
        print (t*n)
        print (i*n)
        print (c*n)
        print (a*n)
```

```python
# Задание 3
n = input()
print('Третий символ строки: ', end='')
print(n[2])
print('Предпоследний символ строки: ', end='')
print(n[-2])
print('Первые пять символов строки: ', end='')
print(n[:5])
print('Вся строка, кроме последних двух символов: ', end='')
print(n[:-2])
print('Все символы строки с четными индексами: ', end='')
print(n[::2])
print('Все символы строки с нечетными индексами: ', end='')
print(n[1::2])
print('Все символы строки в обратном порядке: ', end='')
print(n[::-1])
print('Все символы строки через один в обратном порядке, начиная с последнего : ', end='')
print(n[::-2])
print('Длина всей строки: ', end='')
print(len(n))
```

```python
# Задание 4 

# Для YES
n = str.lower(input()) #.lower игнорирует регистр
q = n[::-1]
if q == n:
  print("YES")
else:
  print("NO")

# Для NO
n = str.lower(input()) #.lower игнорирует регистр
q = n[::-1]
if q == n:
  print("YES")
else:
  print("NO")
```
