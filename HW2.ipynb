Необходимо выделить все биграммы в этом тексте и сохранить их список в какую-либо переменную. Биграммы -- это пары стоящих рядом слов, например, "варкалось хливкие", "хливкие шорьки" и т.д.

```python
# задание 1
text = "варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг"
text = text.split() # разделяем текст на слова
digrams = [] # делаем пустой список
for i in range (len(text) - 1): # проходит цикл по всем словам (их индексам)
    digram = text [i] + " " + text[i+1] # задаем биграмму 
    print (digram) # выводим биграммы
    digrams.append(digram) # добавляем в список 
    
```

**Удивительная кошка**

*Несчастная кошка порезала лапу-

    1. Скопировать текст, сохранить его в переменную.
    2. Узнать длину текста в символах.
    3. Очистить текст от пунктуации, используя list compehensions. На выходе должен получиться список слов.
    4. Объединить слова из получившегося списка в строку через пробел.
    5. С помощью list comprehensions проверить, написано ли слово с большой буквы, и если да, то привести его к нижнему регистру. На выходе должен получиться список из всех слов стихотворения, написанных с маленькой буквы (не только те, которые изначально были с маленькой буквы, а все).

    7. Узнать количество уникальных слов.
    8. Распечатать последние 10 слов.
    9. Соединить полученный в пункте 3 список в строку. Слова в строке должны быть написаны через знак переноса строки.
    10. Распечатать с 20 по 30 символ этой строки (включительно).
    11. Разбить текст на биграммы, записать их в переменную и посчитать частотность каждой биграммы. Биграммы должны быть в виде списка кортежей.

```python
# задание 3

# Скопировать текст, сохранить его в переменную.

text = """
Удивительная кошка

Несчастная кошка порезала лапу-
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!
"""

# Узнать длину текста в символах.

print ('Длина текста в символах:', len(text))

# Очистить текст от пунктуации, используя list compehensions. На выходе должен получиться список слов.

clean_text = [word.strip("!?.,-") for word in text.split()] # в .strip избавляемся от знаков пунктуации, которые есть в тексте
print (clean_text)

# Объединить слова из получившегося списка в строку через пробел

string = ' '.join(clean_text) # превращение списка в строку с пробелами между элементами списка
print (string)

# С помощью list comprehensions проверить, написано ли слово с большой буквы, и если да, то привести его к нижнему регистру. На выходе должен получиться список из всех слов стихотворения, написанных с маленькой буквы (не только те, которые изначально были с маленькой буквы, а все).

lower = ([word.lower() if word.istitle() else word for word in clean_text]) # если слово начинается с большой буквы, то становится с нижним регистром и выводится список всех слов, начинающихся с нижнего регистра 
print (lower)

# Узнать длину текста в словах.

print ('Длина текста в словах:', len(clean_text)) # число элементов в тексте без пунктуации

# Узнать количество уникальных слов.

unique = set(lower) # превращаем список слов с маленькой буквы, в множество, в котором не может быть 2 одинаковых элемента
print ('Уникальных слов:', len(unique)) # выводим количество элементов множества

# Распечатать последние 10 слов. (беру из списка слов без пунктуации, а не изначального текста)

print ('Последние 10 слов:', clean_text[-10:])

# Соединить полученный в пункте 3 список в строку. Слова в строке должны быть написаны через знак переноса строки.

string1 = '\n'.join(clean_text) # \n - знак переноса строки
print (string1)

# Распечатать с 20 по 30 символ этой строки (включительно).

print (string1[19:30])

# Разбить текст на биграммы, записать их в переменную и посчитать частотность каждой биграммы. Биграммы должны быть в виде списка кортежей.

digrams = []
for i in range (len(clean_text) - 1): # цикл
    digram = tuple((clean_text[i], clean_text[i+1])) # делаем биграмму кортежем
    digrams.append(digram) # без списка не получится далее посчитать частотность 
    print (digram)

frequency = set(digrams) # для выяснения частоты, делаем множество из списка биграмм
for digram in frequency:
    print (str(digrams.count(digram)), 'раз') # выводит сколько раз встречается та или иная соответствующая биграмма
```
